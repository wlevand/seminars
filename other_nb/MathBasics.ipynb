## Linear algebra

```python
# from https://gist.github.com/KMChris/8fd878826453c3d55814b3293c7b084c
# https://www.overleaf.com/learn/latex/Matrices

from IPython.display import display, Math

def print_matrix(array, notation='p'):
    matrix = ''
    for row in array:
        try:
            for number in row:
                matrix += f'{number}&'
        except TypeError:
            matrix += f'{row}&'
        matrix = matrix[:-1] + r'\\'
        
    display(Math(fr'\begin{{{notation}matrix}}'+matrix+fr'\end{{{notation}matrix}}'))
```

```python
import numpy as np
x = np.linspace(-10,10,5000)
deltax = x[1]-x[0]

def norm(phi):
    norm = np.sum(np.square(np.abs(phi)))*deltax
    return phi/np.sqrt(norm)

def complex_plot(x,y,prob=True,**kwargs):
    real = np.real(y)
    imag = np.imag(y)
    a,*_ = plt.plot(x,real,label='Re',**kwargs)
    b,*_ = plt.plot(x,imag,label='Im',**kwargs)
    plt.xlim(-2,2)
    #plt.show()
    if prob:
        p,*_ = plt.plot(x,np.abs(y),label=r'$\sqrt{P}$')
        #plt.show()
        return a,b,p
    else:
        return a,b
    
def wave_packet(pos=0,mom=0,sigma=0.2):
    return norm(np.exp(-1j*mom*x)*np.exp(-np.square(x-pos)/sigma/sigma,dtype=complex))
```
